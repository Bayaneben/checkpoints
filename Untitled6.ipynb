# Feature selection

- Id and date are not important to predict price so we are discarding it 
- From the correlation map :  the zipcode, condition, sqft_lot15, and sqft_lot have a low correlation with the price (correlation value less than 0.1 and greater than -0.1 ),  thus, they are not of any significance in predicting price

```python
data = df[['price','sqft_living','grade', 'sqft_above', 'sqft_living15','bathrooms','view',
                     'sqft_basement','bedrooms','floors','waterfront','yr_built','yr_renovated','lat','long']]
data
```

# SPLITTING DATA INTO TRAINING AND TESTING SET

splitting dataset into 20% of test data and 80% for training the model

```python
features=['sqft_living','grade', 'sqft_above', 'sqft_living15','bathrooms','view','sqft_basement','bedrooms','floors','waterfront','yr_built','yr_renovated','lat','long']
X = data[features]
Y = data['price']
```

# Applying linear regression model (example : sqft_living)

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train, x_test , y_train, y_test = train_test_split(data['sqft_living'] , Y , test_size= 0.2)
x_train=np.array(x_train).reshape(-1,1)
x_test=np.array(x_test).reshape(-1,1)
y_train=np.array(y_train).reshape(-1,1)
y_test=np.array(y_test).reshape(-1,1)
model = LinearRegression()
model.fit(x_train , y_train)
y_pred = model.predict(x_test)
plt.scatter(x_test,y_test,color="r")
plt.title("Linear Regression")
plt.ylabel("price")
plt.xlabel("sqft_living")
plt.plot(x_test,model.predict(x_test),color="k")
plt.show()
```

In the LinearRegression plot, the red points are the actual outputs where the black line represents the predicted values.
we detect also an outlier (sqft_living=10000)

# Measure the performance of linear regression (example : sqft_living)

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics
print("MSE", mean_squared_error(y_test,y_pred))
print("R squared", metrics.r2_score(y_test,y_pred))
```

# Applying multiple-linear regression

```python
x_train1, x_test1 , y_train1, y_test1 = train_test_split(X , Y , test_size= 0.2)
model.fit(x_train1 , y_train1)
y_pred1 = model.predict(x_test1)
from sklearn.metrics import mean_squared_error
from sklearn import metrics
print("MSE", mean_squared_error(y_test1,y_pred1))
print("R squared", metrics.r2_score(y_test1,y_pred1))
```

We note that the R squared metric has improved from linear to multiple linear regression

# polynomial regression 

```python
from sklearn.preprocessing import PolynomialFeatures
model = LinearRegression()
poly = PolynomialFeatures(degree=2)
poly.fit_transform(x_train)
model.fit(x_train , y_train)
```

```python
poly.fit_transform(x_test)
y_pred = model.predict(x_test)
print("MSE: ", metrics.mean_squared_error(y_test,y_pred))
print("R squared: ", metrics.r2_score(y_test,y_pred))
```

*******Using Multiple features**********

```python
x_train1, x_test1 , y_train1, y_test1 = train_test_split(X , Y , test_size= 0.2)
from sklearn.preprocessing import PolynomialFeatures
model = LinearRegression()
poly = PolynomialFeatures(degree=2)
poly.fit_transform(x_train1)
model.fit(x_train1 , y_train1)
model.fit(x_train1 , y_train1)
y_pred1 = model.predict(x_test1)
poly.fit_transform(x_test1)
y_pred = model.predict(x_test1)
print("MSE: ", metrics.mean_squared_error(y_test1,y_pred1))
print("R squared: ", metrics.r2_score(y_test1,y_pred1))
```

As we can see the complexe model (all the features) has a R-squared (testing) score of 0.68 whether it is for linear or polynomial regression
